```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
len(df)
```

```python
df.head().T
```

```python
df.dtypes
```

```python
total_charges = pd.to_numeric(df.TotalCharges, errors='coerce')
df[total_charges.isnull()][['customerID', 'TotalCharges']]
```

```python
#set missing values in TotalCharges to zero
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df.TotalCharges = df.TotalCharges.fillna(0)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
string_columns = list(df.dtypes[df.dtypes == 'object'].index)

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.churn = (df.churn == 'yes').astype(int)
```

```python
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
```

```python
df_train_full.head()
```

```python
df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=11)

#save the target value from the train/val datasets out to a separate dataframe
y_train = df_train.churn.values
y_val = df_val.churn.values

#del the target variable from both train & val dataframes
del df_train['churn']
del df_val['churn']
```

```python
df_train_full.isnull().sum()
```

```python
#check the distribution of values in the target variable

df_train_full.churn.value_counts()
```

```python
global_mean = df_train_full.churn.mean()
print(round(global_mean,3))
```

```python
#create lists of the categorical & numerical columns
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents', 'phoneservice',
'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup', 'deviceprotection',
'techsupport', 'streamingtv', 'streamingmovies', 'contract', 'paperlessbilling', 'paymentmethod']
numerical = ['tenure', 'monthlycharges', 'totalcharges']

df_train_full[categorical].nunique()
```

```python
#churn rate for males/females only

female_mean = df_train_full[df_train_full.gender == 'female'].churn.mean()
print(round(female_mean, 3))

male_mean = df_train_full[df_train_full.gender == 'male'].churn.mean()
print(round(male_mean, 3))
```

```python
#check churn rate for partner is yes/no
partner_yes = df_train_full[df_train_full.partner == 'yes'].churn.mean()
print(round(partner_yes, 3))

partner_no = df_train_full[df_train_full.partner == 'no'].churn.mean()
print(round(partner_no, 3))
```

```python
df_group = df_train_full.groupby(by='gender').churn.agg(['mean'])
df_group['diff'] = df_group['mean'] - global_mean
df_group['risk'] = df_group['mean'] / global_mean

print(df_group)
```

```python
from IPython.display import display

for col in categorical:
    df_group = df_train_full.groupby(by=col).churn.agg(['mean'])
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['rate'] = df_group['mean'] / global_mean
    display(df_group)
```

```python
from sklearn.metrics import mutual_info_score

def calculate_mi(series):
    return mutual_info_score(series, df_train_full.churn)
    
df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
print(df_mi)
```

```python
#calculating the correlation cofficient

df_train_full[numerical].corrwith(df_train_full.churn)
```

```python
#one hot encoding using sklearn

train_dict = df_train[categorical + numerical].to_dict(orient='records')
```

```python
print(train_dict[0:1])
```

```python
#using sklearn

from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train[0]
```

```python
dv.get_feature_names()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)

```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
y_pred = model.predict_proba(X_val)
```

```python
print(y_pred)
```

```python
#select only the second column of y_pred which is probability of customer will churn
y_pred = model.predict_proba(X_val)[:,1]
```

```python
print(y_pred)
```

```python
#categorize those with a probability of 0.5 or greater as likely to churn
churn = y_pred >= 0.5
```

```python
#compare the original target values of the validation data set to the predictions made on the validation set

comps = (y_val == churn).mean()
```
